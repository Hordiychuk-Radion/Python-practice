```python
df['Location'].value_counts()
```

```python
ax = df.query('Location != "Other" ') \
    .groupby('Location')['Speed_mph'] \
        .agg(['mean', 'count']) \
            .query('count >=10') \
                .sort_values('mean') \
                    .plot(kind ='barh', figsize =(8, 5), title = 'Average Coast Speed by Location')
                    
ax.set_xlabel  ('Average Coaster Speed')    
plt.show()             
```
